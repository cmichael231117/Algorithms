```python
# quick sort
array = [5, 7, 9, 0, 3, 1, 6, 2, 4, 8]

def quick_sort(array, start, end):
    if start >= end:
        return
    pivot = start
    left = start + 1
    right = end
    while left <= right:
        while left <= end and array[left] <= array[pivot]:
            left += 1

        while right > start and array[right] >= array[pivot]:
            right -= 1

        if left > right:
            array[right], array[pivot] = array[pivot], array[right]
        else:
            array[right], array[left] = array[left], array[right]

    quick_sort(array, start, right-1)
    quick_sort(array, right+1, end)
    
quick_sort(array, 0, len(array)-1)
print(array)
```

```python
# quick sort 2
array = [5, 7, 9, 0, 3, 1, 6, 2, 4, 8]

def quick_sort(array):
    
    if len(array)<=1:
        return array

    pivot = array[0]
    tail = array[1:]

    left_side = [x for x in tail if x<=pivot]
    right_side = [x for x in tail if x>pivot]

    return quick_sort(left_side) + [pivot] + quick_sort(right_side)

print(quick_sort(array))
```

```python
def dijkstra(start):
    q = []
    heapq.heappush(q, (0, start))
    distance[start] = 0
    while q:
        dist, now = heapq.heappop(q)
        if distance[now] < dist:
            continue
        for i in graph[now]:
            cost = dist + i[1]
            if cost < distance[i[0]]:
                distance[i[0]] = cost
                heapq.heappush(q, (cost, i[0]))

dijkstra(start)

for i in range(1, n+1):
    if distance[i] == INF:
        print("INFINITY")
    else:
        print(distance[i])
```

```python
n = int(input())

teachers = []
spaces = []

aisle = []
for i in range(n):
    aisle.append(list(input().split()))
    for j in range(n):
        if aisle[i][j] == 'T':
            teachers.append((i, j))
        if aisle[i][j] == 'X':
            spaces.append((i, j))

def check(x, y, direction):
    if direction == 0:
        while x >= 0:
            if aisle[x][y] == 'S':
                return False
            if aisle[x][y] == 'O':
                return True
            x -= 1
    if direction == 1:
        while x < n:
            if aisle[x][y] == 'S':
                return False
            if aisle[x][y] == 'O':
                return True
            x += 1
    if direction == 2:
        while y >= 0:
            if aisle[x][y] == 'S':
                return False
            if aisle[x][y] == 'O':
                return True
            y -= 1
    if direction == 3:
        while y < n:
            if aisle[x][y] == 'S':
                return False
            if aisle[x][y] == 'O':
                return True
            y += 1
    return True


def solution():
    for i in range(len(spaces)):
        x1, y1 = spaces[i]
        aisle[x1][y1] = 'O'
        for j in range(i, len(spaces)):
            x2, y2 = spaces[j]
            aisle[x2][y2] = 'O'
            for k in range(j, len(spaces)):
                x3, y3 = spaces[k]
                aisle[x3][y3] = 'O'
                for teacher in teachers:
                    x, y = teacher
                    if check(x, y, 0) & check(x, y, 1) & check(x, y, 2) & check(x, y, 3):
                        return True
                x3, y3 = spaces[k]
                aisle[x3][y3] = 'X'
            x2, y2 = spaces[j]
            aisle[x2][y2] = 'X'
        aisle[x1][y1] = 'X'
    return False

if solution():
    print('YES')
else:
    print('NO')
```

```python
import heapq
n = int(input())

heap = []
for i in range(n):
    data = int(input())
    heapq.heappush(heap, data)

result = 0

while len(heap) != 1:
    one = heapq.heappop(heap)
    two = heapq.heappop(heap)

    sum_value = one + two
    result += sum_value
    heap.heappush(heap, sum_value)

print(result)
```

```python
def count1(array, x):
    n = len(array)

    a = first(array, x, 0, n-1)

    if a == None:
        return 0
    b = last(array, x, 0, n-1)

    return a - b +1

def first(array, x, start, end):
    if start > end:
        return None
    
    mid = (start + end) / 2

    if (mid == 0 or array[mid-1] < x) and array[mid] == x:
        return mid
    elif (array[mid] >= x):
        return first(array, x, start, mid-1)
    else:
        return first(array, x, mid+1, end)

def last(array, x, start, end):
    if start > end:
        return None
    
    mid = (start + end) / 2

    if (mid == n-1 or array[mid+1] > x) and array[mid] == x:
        return mid
    elif (array[mid] > x):
        return last(array, x, start, mid-1)
    else:
        return last(array, x, mid+1, end)

n, x = map(int, input().split())
array = list(map(int, input().split()))

count = count1(array, x)

if count == 0:
    print(-1)
else:
    print(count)
```

```python
from bisect import bisect_left, bisect_right

def count_by_value(array, left, right):
    left_idx = bisect_left(array, left)
    right_idx = bisect_right(array, right)
    return right_idx - left_idx

```
