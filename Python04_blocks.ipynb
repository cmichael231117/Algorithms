```python
# O(n)
def isPrime(x):
    for i in range(x):
        if x % i == 0:
            return False
    return True

# O(n^(1/2))
import math
def isPrime_half(x):
    for i in range(2, int(math.sqrt(x)) + 1):
        if x % i == 0:
            return False
    return True

# O(nloglogn)
def eratosthenes(n = 1000):
    array = [True for i in range(n+1)]
    array[0] = False
    array[1] = False

    for i in range(2, int(math.sqrt(n) + 1)):
        if array[i] == True:
            j = 2
            while i * j <= n:
                array[i*j] = False
                j += 1

    return array

def intervalSum(m = 5, data = [1,2,3,2,5]):
    count = 0
    interval_sum = 0
    end = 0

    for start in range(len(data)):
        while interval_sum < m and end < len(data):
            interval_sum += data[end]
            end += 1

        if interval_sum == m:
            count += 1

        interval_sum -= data[start]

    return count

def merge(a = [1,3,5], b = [2,4,6,8]):
    n, m = len(a), len(b)
    result = [0] * (n + m)
    i, j, k = 0

    while i < n or j < m:
        if j >= m or (i < n and a[i] <= b[j]):
            result[k] = a[i]
            i += 1
        else:
            result[k] = b[j]
            j += 1
        k += 1

    return result

def partialSum(left = 3, right = 4, data = [10, 20, 30, 40, 50]):
    n = len(data)
    sum_value = 0
    prefix_sum = [0]
    for i in data:
        sum_value += 1
        prefix_sum.append(sum_value)

    return prefix_sum[right] - prefix_sum[left - 1]
```
